```python
import pandas as pd
from prompt_testing import gen_ex

# read the input_text.txt file from the texts folder
with open("prompts/prompt_a.txt", "r") as f:
    prompt_template_a = f.read()

with open("prompts/prompt_b.txt", "r") as f:
    prompt_template_b = f.read()

with open("prompts/prompt_c.txt", "r") as f:
    prompt_template_c = f.read()

with open("prompts/prompt_d.txt", "r") as f:
    prompt_template_d = f.read()

with open("prompts/prompt_e.txt", "r") as f:
    prompt_template_e = f.read()

with open("prompts/prompt_f.txt", "r") as f:
    prompt_template_f = f.read()

prompt_templates = [prompt_template_f]

# read the input_text.txt file from the texts folder
with open("texts/input_text.txt", "r") as f:
    input_text = f.read()

# read the reference_text.txt file from the texts folder
with open("texts/reference.txt", "r") as f:
    reference = f.read()

# read the input_text.txt file from the texts folder
with open("texts/input_text_2.txt", "r") as f:
    input_text_2 = f.read()

# read the reference_text.txt file from the texts folder
with open("texts/reference_2.txt", "r") as f:
    reference_2 = f.read()

# read the input_text.txt file from the texts folder
with open("texts/input_text_3.txt", "r") as f:
    input_text_3 = f.read()

# read the reference_text.txt file from the texts folder
with open("texts/reference_3.txt", "r") as f:
    reference_3 = f.read()

test_cases = [{"input_text": input_text, "reference": reference}, 
              {"input_text": input_text_2, "reference": reference_2},
              {"input_text": input_text_3, "reference": reference_3}]

results = gen_ex(prompt_templates, test_cases, test_runs=10, project_name="ex1-style-transfer")

df = pd.DataFrame(results)
# check if there is already a csv file called "test_results.csv"
import os
if os.path.isfile("test_results.csv"):
    # if there is, load it into a dataframe
    prev_df = pd.read_csv("test_results.csv")
    # add the new results to the dataframe
    df = prev_df.append(df, ignore_index=True)

# save the dataframe to a csv file
df.to_csv("test_results.csv", index=False)

# average the mean embedding distance for each pid
print(df.groupby("pid")["embedding_distance"].mean())
df.tail(10)
```
